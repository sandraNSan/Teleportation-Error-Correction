```python
from qiskit.circuit.library import XGate, ZGate
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit.quantum_info import Statevector, partial_trace, DensityMatrix, state_fidelity, entropy
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
import matplotlib.pyplot as plt
from qiskit.circuit.library import XGate
 
import numpy as np 
```

```python

#---------------------------------------------------------------------------------------------------------------------
# Define registers to create a quantum circuit, with 3 qubits and 2 classical bits 
#
#              qubit_0 ---> U = The unknown state we want to teleport that Alice has 
#              qubit_1 ---> A = Alice's qubit of the Bell pair
#              qubit_2 ---> B = Bob's qubit of the Bell pair 
#              Classical bits ---> C = The qubits after measurement by Alice collapse the global superposition
#----------------------------------------------------------------------------------------------------------------------


Unknown = QuantumRegister (1, "Unknown")
Alice = QuantumRegister (1, "Alice")
Bob = QuantumRegister (1, "Bob")

Classical = ClassicalRegister (2, "Classical bit")

qc = QuantumCircuit (Unknown, Alice, Bob, Classical) # Creates quantum circuit

from qiskit.circuit import Gate

#----------------------------------------------------------------------------------------------------------------------
# Forming a Bell state pair to entangle qubit_1 and qubit_2 so measurements can be performed for teleportation.
#             
#              1. Apply Hadamard gate H on qubit 1 to create a superposition of Alice's pair (not entangled)
#
#              2. Apply CNOT gate to entangle Alice & Bob pair:   Control qubit ---> qubit_1 
#                                                                 Target qubit  ---> qubit_2
#----------------------------------------------------------------------------------------------------------------------

qc.barrier(Unknown[0], Alice[0], Bob[0], label="Bell pair preparation")

qc.h(Alice[0])
qc.cx(Alice[0], Bob[0])


#----------------------------------------------------------------------------------------------------------------------
# Randomly generate a point on the Bloch sphere and assign it to the unknown qubit_0 state to be teleported  
#----------------------------------------------------------------------------------------------------------------------
qc.barrier(Unknown[0], Alice[0], Bob[0], label="Random unknown (input) state")

np.random.seed (50)
theta = np.pi * np.random.uniform (0.0,1.0) # Multiplies a random number between 0 and 1 by pi
phi = np.pi * np.random.uniform (0.0,2.0)

qc.u(theta, phi, 0.0, Unknown[0])


#----------------------------------------------------------------------------------------------------------------------
# Formimg an entangled pair between Alice qubit_1 and the unkown state qubit_0 by a change in basis 
#              
#              1. Apply CNOT gate:   Control qubit ---> qubit_0  iff U = 1 then flips A
#                                    Target qubit  ---> qubit_1
#              2. Apply Hadamard gate H  for a change in basis:  computational basis ---> Bell basis measurement 
#
# Expresses the combined state of U and A in a Bell basis 
#----------------------------------------------------------------------------------------------------------------------

qc.barrier(Unknown[0], Alice[0], Bob[0], label="Entangle input state and Alice's Bell pair")
qc.cx(Unknown[0], Alice[0])
qc.h(Unknown[0])

 
#----------------------------------------------------------------------------------------------------------------------
# Performing a Bell measurement on Alice's qubit_1 and the unknown state qubit_0 
#
#       1. Before measurement by Alice : The system is in a quantum superposition of four branches, such that the 
#                                        unknown state is no longer local but a joint system. 
#       2. Alice makes a measurement   : Collapses the superposition to one definite branch, becoming a classical bit                                   
#
#         
#
#----------------------------------------------------------------------------------------------------------------------

qc.barrier(Unknown[0], Alice[0], Bob[0], label="Bell measurement on input state and Alice's Bell pair")
qc.measure(Alice[0], Classical[1])
qc.measure(Unknown[0], Classical[0])


#----------------------------------------------------------------------------------------------------------------------
# Defines the instructions to Bob to obtain the correct teleported state. Alice sends two classical bits to Bob 
#
# If Alice's measurement collapses to one branch, the classical bits she sends to Bob are: 
#
#                                                                 |10> :  "10" ---> Bob performs X gate on the state
#                                                                 |00> :  "00" ---> Bob has a perfect state
#                                                                 |01> :  "01" ---> Bob performs Z gate on the state
#                                                                 |11> :  "11" ---> Bob performs XZ gate on the state
#
#----------------------------------------------------------------------------------------------------------------------

qc.barrier(Unknown[0], Alice[0], Bob[0], label="Bob's correction")

with qc.if_test((Classical[1], 1)):
    qc.x(Bob[0])
with qc.if_test((Classical[0], 1)):
    qc.z(Bob[0])

 
qc.draw(output = 'mpl', fold =20)
```
